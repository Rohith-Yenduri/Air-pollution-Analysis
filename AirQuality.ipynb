## Correlational Analysis

```python
for col in data.columns:
    if col != 'AQI' and  col!="month":
        sns.regplot(x=col, y='AQI', data=data,line_kws = {'color':'red'})
        c = data['AQI'].corr(data[col])
        plt.title(f'Correlation between {col} and AQI (Correlation Coefficient: {c:.2f})')
        plt.xlabel(col)
        plt.ylabel('AQI')
        plt.show()

    
```

```python
d=data.drop("month",axis=1)
correlation_matrix = d.corr()

plt.figure(figsize=(10, 8))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix')
plt.xticks(rotation=90)
plt.yticks(rotation=0)

plt.show()
```

# Insights on Air Quality and AQI Trends

Based on the data available, there has been a consistent improvement in air quality over time, as indicated by the Air Quality Index (AQI) since 2017. This improvement is a positive sign that efforts to combat air pollution have been effective.

However, a concerning trend emerged in July 2022 when there was a sudden increase in AQI, reaching the highest level recorded since 2017. This indicates a potential deterioration in air quality during that period, which requires immediate attention and action.

The primary factors responsible for this increase in AQI were higher levels of carbon monoxide (CO) and elevated concentrations of fine particulate matter (PM2.5) during the last few months of 2022. Identifying and addressing the sources of these pollutants is crucial for maintaining and further improving air quality.

Measures should be taken to analyze and control emissions from various sources, such as industrial activities, vehicular traffic, and biomass burning, which are known contributors to CO and PM2.5 levels. Implementation of effective policies and regulations will be essential to mitigate these pollutants and ensure sustainable air quality improvement.


# Prevention and Measures


1. Reduce Emissions from Vehicles
Promote Public Transportation: Encourage the use of buses, trains, and other public transit options to reduce the number of vehicles on the road.
Adopt Electric Vehicles: Support the adoption of electric and hybrid vehicles to lower emissions from traditional gasoline-powered cars.
Carpooling and Ridesharing: Encourage carpooling and ridesharing to reduce the number of vehicles on the road.
2. Enhance Industrial Emission Controls
Implement Emission Standards: Enforce strict emission standards for industries to limit harmful pollutants released into the atmosphere.
Promote Cleaner Technologies: Encourage industries to adopt cleaner technologies and practices to minimize their environmental impact.
3. Reduce Indoor Air Pollution
Ventilation and Air Filters: Ensure proper ventilation in indoor spaces and use air filters to remove indoor air pollutants.
Avoid Smoking Indoors: Prohibit smoking indoors to prevent exposure to secondhand smoke.
4. Encourage Renewable Energy
Transition to Renewable Sources: Promote the use of renewable energy sources such as solar, wind, and hydroelectric power to reduce reliance on fossil fuels.
Incentives for Renewable Energy: Offer financial incentives and tax breaks to individuals and businesses adopting renewable energy solutions.
5. Afforestation and Green Spaces
Plant Trees: Increase green spaces and plant trees in urban areas to absorb pollutants and improve air quality.
Urban Planning: Incorporate green spaces and parks into urban planning to promote healthier living environments.
6. Waste Management
Reduce Waste Burning: Discourage the burning of waste, as it releases harmful pollutants into the air.
Promote Recycling: Encourage recycling and proper waste disposal to minimize the environmental impact of waste.
7. Education and Awareness
Public Awareness Campaigns: Conduct public awareness campaigns to educate people about the importance of air quality and how they can contribute to improving it.
School Programs: Introduce air quality education in schools to raise awareness among the younger generation.
8. Stricter Regulations and Enforcement
Government Policies: Implement and enforce stringent air quality regulations and standards.
Monitoring and Enforcement: Establish monitoring systems and penalties for non-compliance to ensure industries and individuals adhere to air quality regulations.


# Conclusion

The analysis of air quality in Hyderabad City reveals both improvements and areas of concern. While PM10, PM2.5, and CO levels indicate potential health risks and the need for continued efforts to reduce pollution, the decreasing trends in PM10 and O3 demonstrate successful initiatives.
